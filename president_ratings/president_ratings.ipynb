```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
clinton_csv = "Resources/clinton.csv"
bush2_csv = "Resources/bush2.csv"
obama_csv ="Resources/obama.csv"
trump_csv = "Resources/trump.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
clinton_df = pd.read_csv(clinton_csv, encoding="ISO-8859-1")
bush2_df = pd.read_csv(bush2_csv, encoding="ISO-8859-1")
obama_df = pd.read_csv(obama_csv, encoding="ISO-8859-1")
trump_df = pd.read_csv(trump_csv, encoding="ISO-8859-1")
```

```python
# Show just the header
clinton_df.head()
#bush2_df.head()
#obama_df.head()
#trump_df.head()

```

```python
#Adding President column
clinton_df['President'] = 'Clinton'
bush2_df['President'] = 'George Bush'
obama_df['President'] = 'Barack Obama'
trump_df['President'] = 'Donald Trump'

clinton_df.head()
```

```python
union_df = pd.concat([clinton_df,bush2_df, obama_df, trump_df],ignore_index=True)
```

```python
#Adding Year Column
union_df["End Date"] = union_df["End Date"].str.slice(start=6)
```

```python
union_df.astype({'End Date': 'int32'}).dtypes

arranged_df = union_df.sort_values(by=['End Date'])

arranged_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
arranged_df.to_csv("Output/president_rating.csv", index=False, header=True)
```
