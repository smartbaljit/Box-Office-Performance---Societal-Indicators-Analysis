```python
# convert year columns in both DFs to object
sentimenDF.pub_date=sentimenDF.pub_date.astype('str')
mergedDF.Year=mergedDF.Year.astype('str')
```

```python
sentimenDF=sentimenDF.rename(columns={'pub_date':'Year'})
```

```python
sentimenDF.head()
```

```python
# merge sentiment scores with mergedDF
mergedDF = mergedDF.merge(sentimenDF, on='Year',how='left')
mergedDF.head()
```

```python
#remove extra column
mergedDF=mergedDF.iloc[:,1:]
mergedDF.head()
```

```python
# check row count
mergedDF.count()
```

```python
# output to csv
mergedDF.to_csv('mergedData.csv')
```
