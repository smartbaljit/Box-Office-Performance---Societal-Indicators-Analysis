```python
#This script has to be executed after executing the script - NYTimes_News_Data (For both the data sets - 
# years - 1995-2005 and 2006-2015)

# import dependencies
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint

# Library for sentiment analysis
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score
```

```python
#test vader sentiment analysis function
sentiment_scores("ice cream is great")
```

```python
# Data from 1995-2005

nineties_df = pd.read_csv("NYTimesNewsData_1995_2005.csv", encoding="ISO-8859-1")
```

```python
nineties_df.head()
```

```python
nineties_df.count()
```

```python
#Since the csv is too big with more than 60000 records, skipping 20 rows every time to speed up the execution
new_df= nineties_df.iloc[::20, :]
new_df.head()
```

```python
new_df.count()
```

```python
# add new columns sentiment_score and sentiment to the dataframe

nineties_df['sentiment_score'] = ''
nineties_df['sentiment'] = ''

#loop through each article description and get sentiment info
# Since we are looping through more than 60000 rows and getting sentiments, this will need atleast 30 minutes to execute

for i, row in new_df.iterrows():
        print("Executing row:" +str(i))
        sentiment_score = float(sentiment_scores(row[0])['compound']) # take sentiment of description of each article
        if(sentiment_score >= 0.05):
            new_df.loc[i,'sentiment'] = 'positive'
        elif(sentiment_score <= 0.05):
            new_df.loc[i,'sentiment'] = 'negative'
        else:
            new_df.loc[i,'sentiment'] = 'neutral'
            #print(row)
        new_df.loc[i,'sentiment_score'] = sentiment_score
       
```

```python
new_df.head()
```

```python
# Reading news from the second set of data -  CSV for 2006 to 2015 news
TwoThousands_df = pd.read_csv("NYTimesNewsData_2006-2015.csv", encoding="ISO-8859-1")
```

```python
#Since the file is too large, skipping every 30 records to speed up the execution
CleanedTwoThousands_df= TwoThousands_df.iloc[::30, :]
```

```python
CleanedTwoThousands_df.count()
```

```python
# add new columns sentiment_score and sentiment to the dataframe
CleanedTwoThousands_df['sentiment_score'] = ''
CleanedTwoThousands_df['sentiment'] = ''
```

```python
#loop through each article description and get sentiment info
# Since we are looping through more than 60000 rows and getting sentiments, this will need atleast 30 minutes to execute

for i, row in CleanedTwoThousands_df.iterrows():
    print("Executing row" +str(i))
    try:
        sentiment_score = float(sentiment_scores(row[0])['compound']) # take sentiment of description of each article
        if(sentiment_score >= 0.05):
            CleanedTwoThousands_df.loc[i,'sentiment'] = 'positive'
        elif(sentiment_score <= 0.05):
            CleanedTwoThousands_df.loc[i,'sentiment'] = 'negative'
        else:
            CleanedTwoThousands_df.loc[i,'sentiment'] = 'neutral'
        CleanedTwoThousands_df.loc[i,'sentiment_score'] = sentiment_score
        #print(CleanedTwoThousands_df.loc[i,'sentiment'])
    except: 
        print("An exception occurred")
```

```python
Ninety_df = new_df[['pub_date', 'sentiment', 'sentiment_score']]
Ninety_df
```

```python
CleanedTwoThousands_df = CleanedTwoThousands_df.rename(columns={"2006": "pub_date"})
```

```python
TwoThousands_df1 = CleanedTwoThousands_df[['pub_date', 'sentiment', 'sentiment_score']]
TwoThousands_df1
```

```python
# Export the news data of the year 1995-2005 along with sentiments to .csv file
Ninety_df.to_csv("sentiment_results_Nineties.csv", index=False, header=True)
```

```python
# Exporting the news data of the year 2006-2015 along with sentiments to .csv file
TwoThousands_df1.to_csv("sentiment_results_TwoThousands.csv", index=False, header=True)
```

```python
#Concatenate both the dataframes into a single dataframe

frames = [Ninety_df, TwoThousands_df1]

result_df = pd.concat(frames)
result_df
```

```python
result_df.dtypes
```

```python
#Since the data type of sentiment_score is object, we need to convert it into numeric
result_df['sentiment_score'] = pd.to_numeric(result_df['sentiment_score'], errors='coerce')
```

```python
result_df.dtypes
```

```python
# Getting average of the sentiment score per year
sentiment_avg =result_df.groupby('pub_date')['sentiment_score'].mean()
sentiment_avg
```

```python
summary_Table=pd.DataFrame({"Sentiment Average per year":sentiment_avg})
summary_Table
```

```python
# Save the average sentiment results of all 20 years into a csv file
summary_Table.reset_index().to_csv('SentimentResults_1995_2015.csv')
```
