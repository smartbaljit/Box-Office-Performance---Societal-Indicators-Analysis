```python
import pandas as pd
import os
from datetime import datetime as dt
```

```python
nflx_csv = os.path.join("stock_price_data", "NFLX.csv")
nflx_df = pd.read_csv(nflx_csv)
nflx_df = nflx_df.drop(["Open", "High", "Low", "Close", "Volume"], axis=1)
nflx_df = nflx_df.rename(columns={"Adj Close": "NFLX Adj Close"})

nflx_df["Month_Year"] = None
for i, row in nflx_df.iterrows():
    date_full = dt.strptime(row["Date"], "%Y-%m-%d")
    date_year = date_full.year
    date_month = date_full.month
    date_my = dt(date_year, date_month, 1)
    date_my = date_my.strftime("%m/%Y")
    nflx_df.loc[i, "Month_Year"] = date_my

gspc_csv = os.path.join("stock_price_data", "GSPC.csv")
gspc_df = pd.read_csv(gspc_csv)
gspc_df = gspc_df.drop(["Open", "High", "Low", "Close", "Volume"], axis=1)
gspc_df = gspc_df.rename(columns={"Adj Close": "GSPC Adj Close"})

gspc_df["Month_Year"] = None
for i, row in gspc_df.iterrows():
    date_full = dt.strptime(row["Date"], "%Y-%m-%d")
    date_year = date_full.year
    date_month = date_full.month
    date_my = dt(date_year, date_month, 1)
    date_my = date_my.strftime("%m/%Y")
    gspc_df.loc[i, "Month_Year"] = date_my
```

```python
nflx_df_mean = nflx_df.groupby(['Month_Year'])['NFLX Adj Close'].mean().to_frame().reset_index()
nflx_df_mean.head()
```

```python
gspc_df_mean = gspc_df.groupby(['Month_Year'])['GSPC Adj Close'].mean().to_frame().reset_index()
gspc_df_mean.head()
```

```python
movie_merged_csv = os.path.join("mergedData.csv")
movie_merged_df = pd.read_csv(movie_merged_csv, index_col=0)
movie_merged_df.shape
```

```python
movie_merged_df = pd.merge(movie_merged_df, nflx_df_mean, on="Month_Year", how="left")
movie_merged_df.head()
```

```python
movie_merged_df = pd.merge(movie_merged_df, gspc_df_mean, on="Month_Year", how="left")
movie_merged_df.head()
```

```python
movie_merged_df.to_csv(movie_merged_csv)
```
