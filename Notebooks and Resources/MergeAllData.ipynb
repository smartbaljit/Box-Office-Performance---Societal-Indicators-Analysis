```python
# import cleaned up movie data
movieDF = pd.read_csv('Inflation_Adjusted_Movie.csv')
movieDF.dtypes
```

```python
# rename TIME
econDF=econDF.rename(columns={'TIME':'Year'})

# create rolling average columns
econDF['GDP_AVG']=(econDF.GDP+econDF.GDP_1+econDF.GDP_2)/3
econDF['Unemployment_AVG']=(econDF.Unemployment+econDF.Unemployment_1+econDF.Unemployment_2)/3
econDF['Confidence_AVG']=(econDF.Confidence+econDF.Confidence_1+econDF.Confidence_2)/3
econDF['Median_Income_AVG']=(econDF.Median_Income+econDF.Median_Income_1+econDF.Median_Income_2)/3
econDF.head()
```

```python
# join on year
mergedDF = movieDF.merge(econDF, on='Year',how='left')
```

```python
# bring in presidential rating data
presDF = pd.read_csv('president_ratings/Output/president_rating.csv')
presDF.head()
```

```python
presDF.dtypes
```

```python
# create month year column for presDF
presDF['Month_Year'] = presDF['Start Date'].astype('datetime64').dt.strftime('%m/%Y')
presDF.head()
```

```python
presDF['Start Date'].head()
```

```python
# There are multiple surveys taken per month per president, will need to average out ratings to avoid duplications when joining
# create group by DF by Month_Year and President

presGroupDF = presDF.groupby(by=['Month_Year','President']).mean().reset_index()
presGroupDF.head(20)
```

```python
# for cases where a Presidency was in transition in January, take the first president
for key, value in presGroupDF.iterrows():
    if str(list(value)[0])[:2]=='01':
        print(f'index:{key} - {list(value)[0]} - {list(value)[1]}')
```

```python
# remove 01-2009 for Barack Obama
presGroupDF.loc[(presGroupDF.Month_Year == '01/2009')&(presGroupDF.President == 'Barack Obama')]
```

```python
# remove 01-2017 for Donald Trump
presGroupDF.loc[(presGroupDF.Month_Year == '01/2017')&(presGroupDF.President == 'Donald Trump')]
```

```python
presGroupDF.drop([16,26])
```

```python
# create month year column for merged DF
mergedDF['Month_Year'] = mergedDF['Release Date'].astype('datetime64').dt.strftime('%m/%Y')
mergedDF.head()
```

```python
# merge presidential ratings with merged DF based on month year column
mergedDF = mergedDF.merge(presGroupDF, on='Month_Year',how='left')
mergedDF.head()
```
